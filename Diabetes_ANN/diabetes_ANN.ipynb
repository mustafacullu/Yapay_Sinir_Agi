```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

```

```python
dataset = pd.read_csv('projedata.txt', delimiter=',')

def pre_process(data):
    data['gender'].replace('Female', 1, inplace=True)
    data['gender'].replace('Male', 2, inplace=True)
    data['gender'].replace('Other', 3, inplace=True)

    data['smoking_history'].replace('never', 1, inplace=True)
    data['smoking_history'].replace('No Info', 2, inplace=True)
    data['smoking_history'].replace('former', 3, inplace=True)
    data['smoking_history'].replace('not current', 4, inplace=True)
    data['smoking_history'].replace('current', 5, inplace=True)
    data['smoking_history'].replace('ever', 6, inplace=True)

pre_process(dataset)

X = dataset.iloc[:, :8].values
y = dataset.iloc[:, 8].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = 0.001  
epochs_num = 30

model = Sequential()
model.add(Dense(6, input_dim=8, activation='tanh'))
model.add(Dense(4, activation = 'tanh'))
model.add(Dense(2, activation = 'tanh'))
model.add(Dense(1, activation='sigmoid'))

optimizer = Adam(learning_rate=lr)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, batch_size=30, epochs=epochs_num)

test_result = model.evaluate(X_test, y_test)
```

```python
for i in range(len(test_result)):
    print(model.metrics_names[i], ':', test_result[i])

new_data_point = np.array([[7, 150, 80, 31, 0, 35.7, 1.130, 30]])
output = model.predict(new_data_point)

print(output)

if output[0, 0] > 0.5:
    print('diyabetli')
else:
    print('diyabetsiz')
```
